```python
native_country=[]
count1=-1
for i in data.native_country:
    count1=count1+1
    if i=="?":
        native_country.append(count1)
```

```python
data = data.drop(['race'],axis='columns')
```

```python
def sex_c(x):
    if x=="Male":
        return 0
    else:
        return 1
```

```python
data.sex=data.sex.apply(sex_c)
```

```python
native_country_dummies = pd.get_dummies(data.native_country)
data = pd.concat([data,native_country_dummies],axis='columns')
data = data.drop(['native_country','United-States'],axis='columns')
```

```python
education_dummies = pd.get_dummies(data.education)
data = pd.concat([data,education_dummies],axis='columns')
data = data.drop(['10th','education'],axis='columns')
```

```python
marital_status_dummies = pd.get_dummies(data.marital_status)
data = pd.concat([data,marital_status_dummies],axis='columns')
data = data.drop(['Divorced','marital_status'],axis='columns')
```

```python
occupation_dummies = pd.get_dummies(data.occupation)
data = pd.concat([data,occupation_dummies],axis='columns')
data = data.drop(['Tech-support','occupation'],axis='columns')
```

```python
relationship_dummies = pd.get_dummies(data.relationship)
data = pd.concat([data,relationship_dummies],axis='columns')
data = data.drop(['Unmarried','relationship'],axis='columns')
```

```python
def numeric_income(x):
    if x=="<=50K":
        return 0
    else:
        return 1
```

```python
workclass_dummies = pd.get_dummies(data.workclass)
data = pd.concat([data,workclass_dummies],axis='columns')
data = data.drop(['Private','workclass'],axis='columns')
```

```python
data.income=data.income.apply(numeric_income)
```

```python
X=data.drop(['income'],axis='columns')
```

```python
Y=data.income
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=.7)
```

```python
LogReg = LogisticRegression()
```

```python
LogReg.fit(X_train,Y_train)
print("TRAIN ACCURACY:",LogReg.score(X_train,Y_train))
```

```python
print("TEST ACCURACY:",LogReg.score(X_test,Y_test))
```
